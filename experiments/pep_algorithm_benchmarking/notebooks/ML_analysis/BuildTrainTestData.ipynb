```python
from pathlib import Path

import numpy as np
import pandas as pd

%matplotlib widget
%load_ext autoreload
%autoreload 2
```

```python
root_path = Path("../../")
```

```python
result_path = root_path.joinpath("results")
result_path
```

```python
empkins_pivot_df = pd.read_csv(
    result_path.joinpath("pivot_dataframe_b_point/empkins_pivot_dataframe_b_point_algos_ms.csv")
).drop(columns="Unnamed: 0")
guardian_pivot_df = pd.read_csv(
    result_path.joinpath("pivot_dataframe_b_point/guardian_pivot_dataframe_b_point_algos_ms.csv")
).drop(columns="Unnamed: 0")
```

#### Combine the empkins and guardian dataframes

```python
combined_df = pd.concat([empkins_pivot_df, guardian_pivot_df])
combined_df
```

```python
print(
    f"Length of empkins dataframe: {empkins_pivot_df.shape[0]} and length of guardian dataframe: {guardian_pivot_df.shape[0]} should add up: {empkins_pivot_df.shape[0] + guardian_pivot_df.shape[0]} to the length of the combined dataframe: {combined_df.shape[0]}"
)
```

#### Replace the NaN values in the condition column with another meaningful expression

```python
combined_df["condition"] = combined_df["condition"].replace(np.nan, "default")
combined_df
```

#### Check how many nan values are present in the data

```python
null_df = combined_df[combined_df.isnull().any(axis=1)]
print(f"A total of {null_df.shape[0]} rows contain nan values.")
```

#### Drop rows containing nan

```python
clean_combined_df = combined_df.dropna()
clean_combined_df
```

```python
print(
    f"The length of the combined dataframe: {combined_df.shape[0]} minus the amount of nan values: {null_df.shape[0]} equals: {combined_df.shape[0] - null_df.shape[0]} and should match the length of the clean_combined dataframe: {clean_combined_df.shape[0]}."
)
```

#### Split the dataset into train and target

```python
train_data = clean_combined_df.drop(columns=["b_point_samplereference"])
train_data
```

```python
target_data = clean_combined_df[
    ["participant", "condition", "phase", "heartbeat_idreference", "b_point_samplereference"]
]
target_data
```

#### Save the train and target data

```python
train_data.to_csv(result_path.joinpath("train_test_data/train_data.csv"))
target_data.to_csv(result_path.joinpath("train_test_data/target_data.csv"))
```
