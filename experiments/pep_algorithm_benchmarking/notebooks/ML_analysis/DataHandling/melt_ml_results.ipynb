## Dataframe conversion from wide to long format

### Setup and Helper Functions

#### Imports

```python
import pandas as pd
from pathlib import Path
```

#### Datapaths

```python
data_path = Path("../../../results/data/b-point")
```

### Convert merged results from wide to long

```python
merged_ml_results = pd.read_csv(data_path.joinpath("merged_ml_results_b_point_paper_rr_only.csv"), index_col=[0,1,2,3])
merged_ml_results
```

```python
merged_ml_results_long = merged_ml_results.reset_index().melt(id_vars=["participant", "condition", "phase", "heartbeat_id_reference"], var_name="Algorithm", value_name="B-Point")
merged_ml_results_long
```
