```python
import pandas as pd
from pathlib import Path
```

```python
model_path = Path("../../../results/models/b-point")
data_path = Path("../../../results/data/b-point")
```

```python
best_estimatorids = ["MinMax_None_RFR_RR-Interval-Include-Nan", "SS_None_RFR_RR-Interval-Include-Nan", "MinMax_None_DTR_RR-Interval-Include-Nan", "SS_SFM_RFR_RR-Interval-Median-Imputed", "MinMax_KBest_RFR_RR-Interval-Median-Imputed", "MinMax_SFM_RFR_RR-Interval-Median-Imputed"]
```

```python
merged_permuter = pd.read_json(model_path.joinpath(f"permuter_summary/merged_b_point_permuter_paper.json"), orient="records",
                                   lines=True).set_index(['pipeline_scaler', 'pipeline_reduce_dim', 'pipeline_clf'])
```

```python
best_estimator_results = merged_permuter[merged_permuter['EstimatorID'].isin(best_estimatorids)].drop(columns=["conf_matrix_folds", "true_labels", "true_labels_folds", "predicted_labels", "predicted_labels_folds", "train_indices", "train_indices_folds", "test_indices", "test_indices_folds"])
best_estimator_results.sort_values(by=["mean_test_neg_mean_absolute_error"])
```

```python
best_estimator_results = merged_permuter[merged_permuter['Dataset'] == 'RR-Interval-Median-Imputed'].drop(columns=["conf_matrix_folds", "true_labels", "true_labels_folds", "predicted_labels", "predicted_labels_folds", "train_indices", "train_indices_folds", "test_indices", "test_indices_folds"])
best_estimator_results.sort_values(by=["mean_test_neg_mean_absolute_error"])
```
