## Evaluate the performance of the best ML-Pipeline trained on the annotations of rater 1 on the annotaions of rater 2

### Setup and Helper Functions

#### Imports

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

#### Datapaths

```python
data_path = Path("../../../results/data/b-point/")
```

### Load data

```python
rater_02 = pd.read_csv(data_path.joinpath("combined_data_rater_02.csv"), index_col=[0,1,2,3,4])
rater_02.index = rater_02.index.droplevel(0)
ml_results = pd.read_csv(data_path.joinpath("merged_ml_results_b_point_paper_rr_only.csv"), index_col=[0,1,2,3])
intersecting_ids = ml_results.index.intersection(rater_02.index)
```

#### Select intersecting cardiac cycles between rater 1 and rater 2
- The mismatch comes from the fact that some cardiac cycles may have been excluded by rater 1, while rater 2 included them or vice versa (Exclusion criteria may involve severe artefacts)

```python
ml_results.loc[intersecting_ids]
```

```python
rater_02.loc[intersecting_ids]
```

### Join best ML-Pipeline predictions to the dataframe with annotations of rater 2

```python
performance_rater_2 = rater_02.loc[intersecting_ids].join(ml_results.loc[intersecting_ids, ['MinMax_None_RFR_RR-Interval-Include-Nan']])
performance_rater_2
```

### Compute the error of the best ML-Pipeline on the annotations of rater 2

```python
performance_rater_2['ml_prediction_error'] = performance_rater_2['MinMax_None_RFR_RR-Interval-Include-Nan'] - performance_rater_2['b_point_sample_reference']
performance_rater_2['ml_prediction_abs_error'] = np.abs(performance_rater_2['MinMax_None_RFR_RR-Interval-Include-Nan'] - performance_rater_2['b_point_sample_reference'])
performance_rater_2
```
